```python
#import libraries
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
import plotly.express as px
```

```python
# Making column
names = ['Azmiya','Fida','Shuqran']
height = [5.7,5.9,6.2]

# Create a line plot
px.line(x=names,y=height,title='Height of students')
```

## Barchart from the above DATASET

```python
# Making column
names = ['Azmiya','Fida','Shuqran']
height = [5.7,5.9,6.2]

# Create a line plot
px.bar(x=names,y=height,title='Height of students')
```
